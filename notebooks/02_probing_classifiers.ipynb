# Notebook 02: Linear Probing

Trains linear probes for drug mechanisms, target families, etc. Visualizes layer-wise probe accuracy to reveal where biological knowledge emerges.

```python
import sys
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import json
sys.path.insert(0, str(Path.cwd().parent / 'src'))
from activations import ActivationDataset
from probes import BiologicalProbe
import visualization as viz
plt.style.use('seaborn-v0_8-darkgrid')
print('=' * 70)
print('BIOLENS: LINEAR PROBING ANALYSIS')
print('=' * 70)
```

## 1. Load Activation Dataset

```python
print('\n[1/4] Loading activation dataset...')
dataset_path = Path.cwd().parent / 'data' / 'processed' / 'activation_dataset_v1'
dataset = ActivationDataset.load(str(dataset_path))
print(f'✓ Loaded: {dataset.n_prompts} prompts, {len(dataset.layers)} layers')
print(f'  Concepts: {list(dataset.labels.keys())}')
```

## 2. Train Probes for Each Concept

```python
print('\n[2/4] Training linear probes across layers...')
print('(This may take 1-2 minutes)')
all_results = {}
for concept_name in dataset.labels.keys():
    print(f'\n  {concept_name}:')
    probe = BiologicalProbe(concept=concept_name, n_folds=5, max_iter=1000, C=1.0)
    activations_by_layer, labels = dataset.get_labeled_subset(concept_name)
    results = probe.sweep_layers(activations_by_layer, labels)
    all_results[concept_name] = results
```

## 3. Save Probe Results

```python
print('\n[3/4] Saving probe results...')
output_dir = Path.cwd().parent / 'results' / 'probing'
output_dir.mkdir(parents=True, exist_ok=True)
summary = {}
for concept, results in all_results.items():
    summary[concept] = {
        'best_layer': results[np.argmax([r.cv_mean for r in results])].layer,
        'best_accuracy': float(max(r.cv_mean for r in results)),
        'layer_accuracies': {r.layer: {'mean': float(r.cv_mean), 'std': float(r.cv_std)} for r in results},
    }
with open(output_dir / 'probe_summary.json', 'w') as f:
    json.dump(summary, f, indent=2)
print(f'✓ Saved probe summary to {output_dir / "probe_summary.json"}')
```

## 4. Visualization: Layer-wise Probe Accuracy

```python
print('\n[4/4] Generating visualizations...')
figures_dir = Path.cwd().parent / 'results' / 'figures'
figures_dir.mkdir(parents=True, exist_ok=True)
fig = viz.plot_probe_accuracy_by_layer(
    all_results,
    save_path=str(figures_dir / '02_probe_accuracy_by_layer.png'),
    title='Linear Probe Accuracy: Biological Concepts Emerge at Different Layers',
)
print('✓ Saved figure: 02_probe_accuracy_by_layer.png')

n_concepts = len(all_results)
n_layers = len(dataset.layers)
accuracy_matrix = np.zeros((n_concepts, n_layers))
concept_names = list(all_results.keys())
for i, concept in enumerate(concept_names):
    for j, r in enumerate(all_results[concept]):
        accuracy_matrix[i, j] = r.cv_mean

fig, ax = plt.subplots(figsize=(14, 4))
im = ax.imshow(accuracy_matrix, aspect='auto', cmap='viridis', vmin=0.4, vmax=1.0)
ax.set_xticks(range(0, n_layers, 2))
ax.set_xticklabels([f'L{i}' for i in range(0, n_layers, 2)])
ax.set_yticks(range(n_concepts))
ax.set_yticklabels(concept_names)
ax.set_xlabel('Layer')
ax.set_title('Heatmap: Linear Probe Accuracy Across Concepts and Layers')
plt.colorbar(im, ax=ax, label='CV Accuracy')
plt.tight_layout()
plt.savefig(figures_dir / '02_probe_accuracy_heatmap.png', dpi=300, bbox_inches='tight')
print('✓ Saved figure: 02_probe_accuracy_heatmap.png')
plt.show()
```

## 5. Interpretation

```python
print('\n' + '=' * 70)
print('INTERPRETATION')
print('=' * 70)
for concept_name, results in all_results.items():
    accuracies = [r.cv_mean for r in results]
    best_idx = np.argmax(accuracies)
    best_result = results[best_idx]
    print(f'\n{concept_name}:')
    print(f'  ✓ Best layer: {best_result.layer} (accuracy: {best_result.cv_mean:.3f})')
    print(f'  ✓ Early layers (0-5): {np.mean(accuracies[:6]):.3f}')
    print(f'  ✓ Middle layers (6-18): {np.mean(accuracies[6:18]):.3f}')
    print(f'  ✓ Late layers (18+): {np.mean(accuracies[18:]):.3f}')
    emergent_layer = next((i for i, acc in enumerate(accuracies) if acc > 0.6 and (i == 0 or accuracies[i-1] <= 0.6)), None)
    if emergent_layer is not None:
        print(f'  ✓ Concept emerges at layer {emergent_layer}')
    else:
        print(f'  ✓ Concept not clearly decodable across layers')
print('\n' + '=' * 70)
print('✓ Notebook 02 complete!')
print('=' * 70)
```
